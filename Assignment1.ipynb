```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
def tanh(x):
    return np.tanh(x)
def relu(x):
    return np.maximum(0,x)
def leaky_relu(x):
    return np.maximum(0.01*x,x)
```

```python
X=np.linspace(-5,5,100)
```

```python
output_sigmoid=sigmoid(X)
output_tanh=tanh(X)
output_relu=relu(X)
output_leaky=leaky_relu(X)
```

```python
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.plot(X,output_sigmoid,label="Sigmoid")
plt.title("Activation for Sigmoid Function")
plt.legend()

plt.subplot(2,2,2)
plt.plot(X,output_leaky,label="Leaky")
plt.title("Activation for Leaky Function")
plt.legend()

plt.subplot(2,2,3)
plt.plot(X,output_relu,label="Relu")
plt.title("Activation for Relu Function")
plt.legend()

plt.subplot(2,2,4)
plt.plot(X,output_tanh,label="Tanh")
plt.title("Activation for Tanh Function")
plt.legend()

plt.show()
```
